```python
import matplotlib.pyplot as plt

# Data for plotting (example history dictionary)
history = {
    'train_loss': [0.5, 0.4, 0.3],
    'val_loss': [0.6, 0.5, 0.4],
    'train_acc': [0.7, 0.8, 0.9],
    'val_acc': [0.65, 0.75, 0.85]
}

# Creating a figure to host the plots
plt.figure(figsize=(10, 5))  # Wider figure to host two subplots

# Subplot 1: Loss
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.plot(history['train_loss'], label='Training Loss', linestyle='--', color='blue', linewidth=1)
plt.plot(history['val_loss'], label='Validation Loss', linestyle='--', color='red', linewidth=1)
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Subplot 2: Accuracy

plt.plot(history['train_acc'], label='Training Accuracy', color='blue', linewidth=4)
plt.plot(history['val_acc'], label='Validation Accuracy', color='red', linewidth=4)
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Save the figure
plt.savefig('TCC_model_perf.pdf')  # Save as PDF before displaying
plt.show()  # Display the plot
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history['train_loss'], label='Training Loss', linestyle='--', color='blue', linewidth=2, )
plt.plot(history['val_loss'], label='Validation LosTs', linestyle='--', color='red', linewidth=2)

plt.plot(history['train_acc'], label='Training Accuracy', color='blue', linewidth=5)
plt.plot(history['val_acc'], label='Validation Accuracy', color='red', linewidth=5)
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')

plt.legend()
plt.show()

# save plot in pdf
plt.savefig('TCC_model_perf.pdf')
```

```python
plt.figure(figsize=(12, 4))
ax1 = plt.subplot(1, 2, 1)

# Plotting the accuracy on the primary y-axis
ax1.plot(history['train_acc'], label='Training Accuracy', color='blue', linewidth=5)
ax1.plot(history['val_acc'], label='Validation Accuracy', color='red', linewidth=5)
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.tick_params(axis='y')
ax1.set_title('Model Performance')
ax1.legend(loc='upper left')

# Creating a second y-axis for loss
ax2 = ax1.twinx()
ax2.plot(history['train_loss'], label='Training Loss', linestyle='--', color='blue', linewidth=2)
ax2.plot(history['val_loss'], label='Validation Loss', linestyle='--', color='red', linewidth=2)
ax2.set_ylabel('Loss')
ax2.tick_params(axis='y')
ax2.legend(loc='upper right')

plt.show()

# Save the figure
plt.savefig('TCC_model_perf.pdf') 
```
