### Visualization model training

```python
# open history_TCC_inside.pkl'
import pickle
with open('history_TCC_inside_balanced_only_train_eval.pkl', 'rb') as f:
    history = pickle.load(f)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
ax1 = plt.subplot(1, 2, 1)

history['val_loss'] = [x / 100 for x in history['val_loss']]
history['train_loss'] = [x / 100 for x in history['train_loss']]

# Plotting the accuracies
ax1.plot(history['train_acc'], label='Training Accuracy', color='blue', linewidth=2)
ax1.plot(history['val_acc'], label='Validation Accuracy', color='red', linewidth=2)
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.set_title('Accuracy vs. Epoch')
ax1.legend(loc='upper left')

# Right plot for losses
ax2 = plt.subplot(1, 2, 2)
ax2.plot(history['train_loss'], label='Training Loss', linestyle='--', color='blue', linewidth=2)
ax2.plot(history['val_loss'], label='Validation Loss', linestyle='--', color='red', linewidth=2)
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.set_title('Loss vs. Epoch')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.savefig('latex/vis_pdf/TCC_model_perf.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
